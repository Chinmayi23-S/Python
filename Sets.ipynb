## Practice Examples

#### Exercise 1: Removing Duplicates from a List

```python
data_with_duplicates = [1, 5, 2, 8, 3, 5, 1, 9, 2, 7]

unique_elements = list(set(data_with_duplicates))    #Convert list to set using 'set' operation and back to list using 'list' operation
print(f"List after removing duplicates using set : {unique_elements}")
```

#### Exercise 2: Check if user exists in a list

#### Exercise 3: Finding Common Elements (Intersection)

```python
group_a_hobbies = {'reading', 'hiking', 'swimming', 'coding'}
group_b_hobbies = {'hiking', 'gaming', 'swimming', 'cooking'}

common_hobbies = group_a_hobbies.intersection(group_b_hobbies)
print(f"Common Hobbies: {common_hobbies}")
```

#### Exercise 4: Finding Unique Elements Across Collections (Union / Symmetric Difference)

```python
class_a = {'Alice', 'Bob', 'Charlie'}
class_b = {'Charlie', 'David', 'Eve'}

all_students = class_a.union(class_b)
print(f"All students in both classes: {all_students}")

students_only_in_one_class = class_a.symmetric_difference(class_b)
print(f"Students only in one class: {students_only_in_one_class}")
```

#### Exercise 6: Common Hobbies
 Find out which hobbies are common to both sets.

```python
morning_hobbies = {'jogging', 'reading', 'meditation', 'coding'}
evening_hobbies = {'coding', 'reading', 'gaming', 'watching TV'}

common_hobbies = morning_hobbies.intersection(evening_hobbies)
print(f"Common hobbies from both lists : {common_hobbies}")
```

#### Exercise 7: Unique Words
Given a string of text, find all the unique words in it. Ignore case sensitivity and punctuation. You can assume words are separated by spaces for simplicity.

```python
text_string = "Python is fun and Python is powerful"

lower_case = text_string.lower()
split_string = lower_case.split(' ')

set_string = set(split_string)
print(lower_case)
print(split_string)
print(set_string)
```

#### Exercise 8: Check if Subset
Determine if a set A is a subset of set B

```python
set_A = {1, 2, 3}
set_B = {1, 2, 3, 4, 5}
set_C = {1, 6}

print(set_A.issubset(set_B))
print(set_C.issubset(set_A))
```
