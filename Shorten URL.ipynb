```python
import requests
long_url = input("Enter Web URL:")

short_url = f"https://tinyurl.com/api-create.php?url={long_url}"

response = requests.get(short_url)

if response.status_code == 200:
    print("SHORT_URL:", response.text)
else:
    print("Fail to generate short URL")
```
